```python
df = pd.read_excel('qualitative_analysis.xlsx')

def remove_edge_cases(r):
    
    if isinstance(r, str) and r.startswith('EDGE'):
        return None
    return r

df = df.applymap(remove_edge_cases)

df.head()
```

```python
ddf = pd.DataFrame.from_dict(
    {'adaptcentre': [
        df['adaptcentre_info.added'].notnull().sum(),
        df['adaptcentre_info.removed'].notnull().sum(),
        df['adaptcentre_info.repeated'].notnull().sum()
    ],
     'upf-forge': [
        df['upf_forge.added'].notnull().sum(),
        df['upf_forge.removed'].notnull().sum(),
        df['upf_forge.repeated'].notnull().sum()
    ],
     'template-pipe-multi': [
        df['me.added'].notnull().sum(),
        df['me.removed'].notnull().sum(),
        df['me.repeated'].notnull().sum()
    ],
    }, orient='index', columns=['added', 'removed', 'repeated'])

ddf['total'] = ddf.sum(axis=1)
```
