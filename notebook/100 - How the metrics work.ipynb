```python
import os
import pandas as pd
pd.set_option('display.max_colwidth', 400)
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
def pretty_print(df):
    return display( HTML( df.to_html().replace("\\n","<br>") ) )
import glob
os.sys.path.insert(0, '../evaluation')
%matplotlib inline
```

```python
models = [os.path.basename(p) for p in glob.glob(f'../data/models/how_metrics_work/*')]

dfs = [pd.read_csv(f'../data/models/how_metrics_work/{model}/system_evaluation.csv', 
                   index_col=['subset', 'references', 'metric']) for model in models]

df = pd.concat(dfs, keys=models)
df = df.unstack().reset_index(level=[1, 2], drop=True)[[('value', 'bleu'), ('value', 'meteor'), ('value', 'ter')]].sort_values(('value', 'bleu'), ascending=False)
df
```
